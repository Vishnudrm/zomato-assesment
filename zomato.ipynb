```python
#reading and displaying the csv files
```

```python
df=pd.read_csv('zomatof.csv')
df.head()
```

```python
#getting the number of rows and columns in the dataset
```

```python
df.shape
```

```python
df.columns
```

```python
#dropping the unwanted data
```

```python
df = df.drop(['url', 'address', 'phone', 'menu_item', 'dish_liked', 'reviews_list'], axis=1, errors='ignore')
```

```python
df.head()
```

```python
# getting information about the data
```

```python
df.info()
```

```python
#dropping the duplicate values of rate
```

```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
df['rate'].unique()
```

```python
#deleting string values and also the "/" symbol from the rates
```

```python
def handlerate(value):
    if (value=='NEW'or value=='-'):
        return np.nan
    else:
        value=str(value).split('/')
        value=value[0]
        return float(value)
df['rate']=df['rate'].apply(handlerate)
df['rate'].head()
```

```python
#finding if there is any null values
```

```python
df.rate.isnull().sum()
```

```python
#filling the null values in the rate with the mean
```

```python
df['rate'].fillna(df['rate'].mean(),inplace =True)
df['rate'].isnull().sum()
```

```python
df.info()
```

```python
#dropping the null values
```

```python
df.dropna(inplace=True)
df.head()
```

```python
#renaming the columns 
```

```python
df.rename(columns={'approx_cost(for two people)':'cost2plates','listed_in(type)':'type'},inplace=True)
df.head()
```

```python
#finding the unique values
```

```python
df['location'].unique()
```

```python
df['listed_in(city)'].unique()
```

```python
#as both location and listed in are same delete any one
```

```python
df=df.drop(['listed_in(city)'],axis=1,errors='ignore')
```

```python
df.head()
```

```python
df['cost2plates'].unique()
```

```python
#removing the ,from cost2plates
```

```python
def handlecomma(value):
    value=str(value)
    if ',' in value:
        value =value.replace(',','')
        return float(value)
    else:
        return float(value)
df['cost2plates']=df['cost2plates'].apply(handlecomma)
df['cost2plates'].unique()
```

```python
df.head()
```

```python
#cleaning rest type 
```

```python
df['rest_type'].value_counts()
```

```python
rest_types=df['rest_type'].value_counts(ascending =False)
rest_types
```

```python
#creating a function and storing less number of rest_type into a variable "others"
```

```python
rest_types_lessthan1000=rest_types[rest_types<1000]
rest_types_lessthan1000
```

```python
#function
```

```python
def handle_rest_type(value):
    if(value in rest_types_lessthan1000):
        return 'others'
    else:
        return value
df['rest_type']=df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

```python
df.head()
```

```python
#cleaning location column
```

```python
df['location'].value_counts
```

```python
location=df['location'].value_counts(ascending=False)
location_lessthan300=location[location<300]

def handle_location(value):
    if(value in location_lessthan300):
        return 'others'
    else:
        return value

df['location']=df['location'].apply(handle_location)
df['location'].value_counts()
```

```python
df.head()
```

```python
#cleaning cuisines column
```

```python
df['cuisines'].value_counts()
```

```python
cuisines=df['cuisines'].value_counts(ascending=False)
cuisines_lessthan100=cuisines[cuisines<100]

def handle_cuisines(value):
    if(value in cuisines_lessthan100):
        return 'others'
    else:
        return value

df['cuisines']=df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```
